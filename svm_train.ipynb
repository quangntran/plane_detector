```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt 
%matplotlib inline
from sklearn.metrics import confusion_matrix
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import Perceptron
plt.rcParams.update({'font.size': 13})
# Adapted from: https://github.com/jtsulliv/ML-from-scratch/blob/master/Neural-Networks/perceptron.ipynb

# synthetic data
obs = 26000 # number of observations
class_zeros = np.random.multivariate_normal([0,0], [[1,0.7],[0.7,1]], obs)
class_ones = np.random.multivariate_normal([0,2], [[1,0.7],[0.7,1]], obs)

# generating a column of ones as a dummy feature to create an intercept
# intercept = np.ones((2*obs,1))

# vertically stacking the two classes 
features = np.vstack((class_zeros, class_ones)).astype(np.float32)

# creating the labels for the two classes
label_zeros = np.zeros((obs,1))
label_ones = np.ones((obs,1))

# stacking the labels
labels = np.vstack((label_zeros,label_ones))
labels = labels[:,0]

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.5)

# scatter plot to visualize the two classes
fig, ax = plt.subplots()
ax.scatter(class_zeros[:,0],class_zeros[:,1], label=0,alpha=0.5)
ax.scatter(class_ones[:,0],class_ones[:,1], label=1,alpha=0.3)
ax.legend()
ax.set_xlabel('x1')
ax.set_ylabel('x2')
plt.show()
```

```python
from sklearn import svm
import time
size = 25088
X_train_ = X_train[:size,:]
y_train_ = y_train[:size]
X_test_ = X_test[:size,:]
y_test_ = y_test[:size]

start = time.time()
clf = svm.SVC(kernel='rbf', gamma=2)
clf.fit(X_train_, y_train_)

train_time = time.time() - start
print(train_time,'s')


```
